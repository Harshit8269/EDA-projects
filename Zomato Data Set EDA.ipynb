```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0 : 'Rating_count'})
```

```python
ratings
```

### Observation
 1. when rating between 4.5 to 4.9 ---> Excellent
 2. when rating between 4.0 to 4.4 ---> Very Good
 3. when rating between 3.5 to 3.9 ---> Good
 4. when rating between 2.5 to 3.4 ---> Average
 5. when rating between 1.8 to 2.4 ---> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating_count' ,data = ratings )
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating_count', palette = ['white','red','orange','yellow','green','green'] , hue = 'Rating color' ,data = ratings)
```

### Observation
1. Not rated count is very high
2. max number of ratings are between 2.5 to 3.4

```python
## countplot 
```

```python
sns.countplot(x = 'Rating color', data = ratings, palette = ['grey','red','orange','yellow','green','green'] )
```
